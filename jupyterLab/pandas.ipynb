```python
desc = df['MON2_TX2_PULSE_DURATION'].describe()
print(desc)
print(desc['max'])
```

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

datetime_format = r'%d.%m.%Y %H:%M:%S'
ts = df['TS'].map(lambda x: datetime.strptime(str(x), datetime_format))

y1 = df['MON2_TX2_PULSE_DURATION']
y2 = df['MON2_TX2_PULSE_RISE_TIME']
y3 = df['MON2_TX2_PEAK_POWER_OUTPUT']

plt1 = ax1.plot(ts, y1)
plt1 = ax1.plot(ts, y2)

plt2 = ax2.plot(ts, y3, color="orange")

ax1.legend()
ax1.grid(True)

ax2.legend()
ax2.grid(True)

#fig.autofmt_xdate()

print('mean=%1.2f, min=%1.2f, max=%1.2f' % (desc['mean'], desc['min'], desc['max']))
plt.show()
```
